### Challenge 1: Largest Palindrome
A palindromic number reads the same both ways. For example, 1234321 is a palindrome. The largest palindrome made from the product of two two-digit numbers is 9009 = 91 × 99. Find the largest palindrome made from the product of two three-digit numbers. Afterward, write a brief explanation walking through your code's logic in markdown.

```python
numbers = []

for x in range(100,1000):
    numbers.append(x)
```

```python
numbers
```

```python
multiply_list = []

def multiply_numbers() :
    for topValue in numbers:
        for bottomValue in range (topValue,100,-1):
            x = topValue * bottomValue
            multiply_list.append(str(x))
```

```python
multiply_numbers()
```

```python
multiply_list
```

```python
pals = []

for x in multiply_list:
    if x == x[::-1]:
        pals.append(x)
```

```python
newest_list = [int(i) for i in pals] 
```

```python
max(newest_list)
```

Then I convert that list to a string and see if the string == reverse order of the string. if yes, I add it to a new list
Then I convert that list back to a list of integers and I find the max value of that list.
