```python
from sklearn.linear_model import LinearRegression
lreg = LinearRegression()

X = df.drop('PRICE', axis=1)
y = df['PRICE']

lreg.fit(X, y)
```

```python
lreg.coef_
# You take these coefficients and multiply them by each of your columns to get your predictions
```

```python
np.sum(X*lreg.coef_, axis=1) + lreg.intercept_
# This is basically how linear regression works
# Neural networks basically do the same thing!
# They just use a loooot more computation
# But if you strip it down, it's basically this
```

```python
X * lreg.coef_
```

```python
X.dot(lreg.coef_)
# 'Activation among neurons' is this operation
# How you multiply datasets by each other
```

```python
X = pd.DataFrame(np.random.normal(0, 0.1, 10000).reshape(1000, 10))
# Stepping through the diagram
```

```python
X.head()
```

```python
coef1 = np.random.normal(0, 0.1, 200).reshape(10, 20)
bias1 = np.random.normal(0, 0.1, 20)
# When you first start these are typically randomly generated close to 0
# There's a bias unit for every single column of output, which is why there's 20
```

```python
X_hidden = X.dot(coef1) + bias1
```

```python
coef2 = np.random.normal(0, 0.1, 20)
bias2 = np.random.normal(0, 0.1)

# For every single coefficient there's a corresponding bias
```

```python
y = X_hidden.dot(coef2)
# This process is called forward propagation
```

```python
# When you're training a neural network you're finding the sets of weights at every single internal error
```

```python
# Neural networks allow you to find nonlinear patterns in nonlinear data
# Linear regression by definition is linear
# In addition to passing it through weights, you have to pass it through an activation function
# The activation function is what makes it non-linear!
```

```python
X_hidden
```

```python
# If you're dealing with a hidden layer in a neural network, you're almost always going to use the same activation function
np.maximum(0, X_hidden)
# This is it!
# Just take negative numbers and set em to 0
# RELU... rectified linear unit
```

```python
def sigmoid(X):
    return (1 / (np.exp(-X)))
```

```python
y = X_hidden.dot(coef2)
y
```

```python
#If you're doing classification...
sigmoid(y)
# Notice how the values change
# The sigmoid function tends to squash values to either 0 or 1 
# SQUASHED
```
