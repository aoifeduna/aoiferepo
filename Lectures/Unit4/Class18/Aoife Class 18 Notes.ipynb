```python
import numpy as np
import pandas as pd
```

```python
seeds = pd.read_csv('/Users/aoifeduna/AoifeRepo/aoiferepo/Lectures/Unit4/data/MNCAATourneySeeds.csv')
results = pd.read_csv('/Users/aoifeduna/AoifeRepo/aoiferepo/Lectures/Unit4/data/MNCAATourneyCompactResults.csv')
```

```python
seeds.head()
```

```python
results.head()
```

```python
seeds['Seed'] = seeds['Seed'].str[1:].str[:2].astype(int)
# Making sure everything is just a number
```

```python
results = results[['Season', 'WTeamID', 'WScore', 'LTeamID', 'LScore']]
```

```python
results.columns = ['Season', 'T1TeamID', 'T1Score', 'T2TeamID', 'T2Score']
```

```python
# Just renaming the columns for clarity
# We're ultimately going to have an entry for the winning team and the losing team
```

```python
results_swap = results[['Season', 'T2TeamID', 'T2Score', 'T1TeamID', 'T1Score']]
# Just reversing the order of the columns
```

```python
results.head()
```

```python
results_swap.head()
# It's just the same thing, but in reverse
```

```python
results_swap.columns = results.columns
```

```python
results.head()
```

```python
results['Result'] = 1
# What are the odds that team one beat team two
# The way this is set up, team 1 ALWAYS beat team two. We moved it so wins were on the left side.
results_swap['Result'] = 0
```

```python
game_data = pd.concat([results, results_swap])
# All this is is both data sets concatenated on top of eachother
game_data
```

```python
game_data.head()
# We recorded two entries for every single NCAA tournament game
```

```python
seeds.head()
```

```python
game_data = game_data.merge(seeds, how='left', left_on=['Season', 'T1TeamID'], right_on=['Season', 'TeamID'])
# We're finding out what the seed was for team 1 by doing this
# Game data is the master data set, so we don't want to lose any of those rows
# We want to keep everything on the other side and merge into it
```

```python
game_data.head()
# The seed represents team ONE seed. So we'll rename.
```

```python
game_data.rename({'Seed': 'T1Seed'}, axis=1, inplace=True)
```

```python
game_data
```

```python
# We don't need the last TeamID
game_data.drop('TeamID', axis=1, inplace=True)
```

```python
game_data.head()
```

```python
game_data = game_data.merge(seeds, how='left', left_on=['Season', 'T2TeamID'], right_on=['Season', 'TeamID'])
```

```python
game_data.rename({'Seed': 'T2Seed'}, axis=1, inplace=True)
```

```python
game_data.head()
```

```python
game_data.drop('TeamID', axis=1, inplace=True)
```

```python
game_data.head()
```

```python
game_data['SeedDiff'] = game_data['T1Seed'] - game_data['T2Seed']
game_data.head()
# You would expect negative values to win and positive values to lose
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier()

X = game_data[['SeedDiff']]
# The difference between the teams playing
# When Scikit learn fits models, it wants X to be two dimensional.
# That's why we have two brackets
y = game_data['Result']
# Whether or not the game was lost

# 1 and 0 represent binary categories.
# You can't win and lose at the same time.
```

```python
rf.fit(X, y)
```

```python
rf.predict(X)
# It's just predicting if the team won or lost
```

```python
rf.score(X, y)
```

```python
rf.predict_proba(X)
# Predicting for the first sample 53% that it's 0, 46% chance it's 1
# Predicting the likelihood of winning
# Basically it's taking a majority vote of every single sample on the leaf
```

For a classification tree:

 - continually find the best value to split on
 - split down until you have your individual leaves
 - prediction: majority vote for all samples
 - probabilities: proportion of each outcome on the leaf
 
For 5 vs 11 seeds:

Total of 1000 samples on all of our trees

850 samples have value of 1
150 samples have value of 0

Prediction of ALL samples will be 1
predict_proba = .15 chance value is 0, .85 chance value is 1
