```python
model_cost = np.sum((df['PRICE'] - df['PREDICTION'])**2)
# Typically you square this
# You normally call this the "cost of your model"
# Basically just taking the difference between the answer, square it, sum it up
# This gives you the summation of the squared error
```

```python
naive_cost = np.sum((df['PRICE'] - df['PRICE'].mean())**2)
# This is our error column
# How far off any individual observation is from the average
# This number is basically the variance
```

```python
1 - (model_cost / naive_cost)
# And this gives you R-squared
# R-squared is unitless
```
