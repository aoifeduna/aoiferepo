```python
import seaborn as sns
sns.catplot(x='Pclass',y='Age', data=df);
# Basic chart for categorical data is the catplot
# Categorical data works better overall within seaborn
```

```python
sns.catplot(x='Pclass',y='Age', kind='swarm',data=df);
# The more frequently a particular value occurs, the more it swells out
# Like a very stylized histogram that you can map to categorical data
```

```python
sns.catplot(x='Pclass',y='Age', kind='box',data=df);
```

```python
sns.catplot(x='Pclass',y='Age', kind='boxen',data=df);
```

```python
sns.catplot(x='Pclass',y='Age', kind='violin',data=df);
```

```python
sns.catplot(x='Pclass',y='Age', kind='point',data=df);
```

```python
sns.catplot(x='Pclass',y='Age', kind='bar',data=df);
```

```python
sns.catplot(x='Pclass',y='Age', kind='swarm', hue='Sex', data=df);
# Adds in yet another layer of data
```

```python
sns.catplot(x='Pclass',y='Age', kind='swarm', col='Sex', data=df);
# Taking a dimension and adding it to a column shell or the row shell
```

```python
sns.lmplot(x='Pclass', y='Survived', data=df);
# Drawing a line of best fit with a 90% confidence interval
```

```python
df['Age'] = df['Age'].fillna(29)
sns.distplot(df.Age);
# You can't have any NA values here
```

```python
sns.jointplot(x='Age', y='Fare', data=df[df.Sex == 'male']);
# Histogram for Age is at the top
# Histogram for Fare is on the right
```
