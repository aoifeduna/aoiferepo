```python
import numpy as np
```

```python
var = np.arange(20).reshape(5, 4)
```

```python
var
# Returns a list of lists
# Each list represents that row
```

```python
var.mean()
# Will perform the entire operation on the whole thing simultaneously.
# Getting the mean of all of the values
```

```python
var.mean(axis=1)
# Axes are important concept
# This is performing an operation across the row
```

```python
var.mean(axis = 0)
# Axis 0 means perform the operation down the column
```

```python
var[2:4, :3]
# You can apply it going down rows and across columns
# Graph rows index positions 2 and 3, grab the first three columns
# You can use this in addition to performing operations
# Using the syntax to apply across multiple dimensions
# Always rows first, then columns
```

```python
df.sample(100)
# Gives you a random sample
# Every single time you run this you'll get a different combination of data
```

```python
# If you just want to grab a specific column you can just treat the label like it were a key in the dictionary

df[['Survived', 'Fare', 'Embarked']]
# To get mulitples you pass them as a list
# You'll get a data frame returned
```

```python
df[['Survived', 'Fare', 'Embarked']][:50]
# This passes in a slice
# This would be the first 50 columns of the data
# If you're grabbing something by its label you're basically doing a dictionary lookup
```

```python
df[:10]
# Grabs the first ten rows
```

```python
df[['Survived', 'Fare', 'Embarked']][:50].mean()
```

```python
df['Fare'] < 35
# When you do this inside pandas it provides the index on the lefthand side and then a bunch of true/false values
```

```python
df[df['Fare'] < 35]
# You can ALWAYS wrap it up in another data frame to get every single row where this condition is true
```

```python
df[df['Fare'] < 35][['Survived', 'Pclass']]
# You're doing the same thing by narrowing this further
# Notice the index is not increasing 1 by 1
# Only the columns where the conditions are true are being selected
```

```python
df[df['Fare'] < 35]['Survived'].mean()
# What was the average fare of the survived column?
```

```python
# Syntax for selecting on more than one condition
df[(df['Fare'] < 35) & (df['Sex'] == 'female')]
# The ampersand means you want to evaluate multiple conditions
# Wrap it all up in a dataframe to get the better visual
```

```python
df[(df['Fare'] < 35) | (df['Sex'] == 'female')]
# The bar is the or operator
```

```python
# People sometimes find it hard to read, so they'll save it as a variable
query = (df['Fare'] < 35) | (df['Sex'] == 'female')
df[query]
```
